```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas.io.formats.excel import ExcelFormatter

def single_post_fetch(url):

    response = requests.get(url)
    htmlcontent = response.content
    soup = BeautifulSoup(htmlcontent, 'html.parser')

    sp_company_logo = []
    sp_main_title = []
    sp_location = []
    sp_vacancies = []
    sp_availability = []
    sp_main_experience = []
    sp_main_region = []
    sp_main_education = []
    sp_title = []
    sp_categories = []
    sp_salaries = []
    sp_age = []
    sp_gender = []
    sp_designation = []
    sp_vacant_positions = []
    sp_how_to_apply_ol = []

    sp_last_date = []
    sp_online_apply_link = []
    sp_address = []
    sp_advertisement_image_link = []
    sp_application_form_link = []
    sp_inner_ad_img = []


    #Post Company Logo
    sin_pos_logo = soup.select('.container-fluid .freelance-image img')[0]
    sp_company_logo.append(sin_pos_logo.get('data-lazy-src'))
    #Post Main Title
    sin_pos_main_title = soup.select('.header-details h1')[0]
    sp_main_title.append(sin_pos_main_title.string)
    # Post Location
    sin_pos_main_location = soup.select('.header-details p')[1]
    sp_location.append(sin_pos_main_location.get_text())
    # Post Vacancies
    sin_pos_main_vacancies = soup.select('.header-details ul li')[0]
    sp_vacancies.append(sin_pos_main_vacancies.get_text())

    # Right Side Details
    # Post Availability
    sin_pos_main_availibility = soup.select('.right-side-detail ul li')[0]
    sp_availability.append(sin_pos_main_availibility.get_text())
    # Post Expererience
    sin_pos_main_experience = soup.select('.right-side-detail ul li')[1]
    sp_main_experience.append(sin_pos_main_experience.get_text())
    # Post Region
    sin_pos_main_region = soup.select('.right-side-detail ul li')[2]
    sp_main_region.append(sin_pos_main_region.get_text())
    # Post Education
    sin_pos_main_education = soup.select('.right-side-detail ul li')[3]
    sp_main_education.append(sin_pos_main_education.get_text())
    # Post Title
    sin_pos_title = soup.select('.stickyrow .stickycol9 h2')[0]
    sp_title.append(sin_pos_title.string)
    # Category
    sin_pos_cat = soup.select('.stickyrow .post-meta div:nth-child(2) span a')[0]
    #print("Categories: "+sin_pos_cat.string)
    sp_categories.append(sin_pos_cat.string)
    # Offered Salary
    sin_pos_salary = soup.select('.stickyrow .ms-4 .key-field')[0]
    sp_salaries.append(sin_pos_salary.string)
    sin_pos_age = soup.select('.stickyrow .ms-4 .key-field')[1]
    sp_age.append(sin_pos_age.string)
    #Gender
    sin_pos_gender = soup.select('.stickyrow .ms-4 .key-field')[2]
    sp_gender.append(sin_pos_gender.string)
    #Designation
    sin_pos_designation = soup.select('.stickyrow .ms-4 .key-field')[4]
    sp_designation.append(sin_pos_designation.get_text())

    #Vaccant Positions
    if "Table of Conte" in str(soup.select('.pcontent')):
        sp_vacant_positions.append(soup.select('.pcontent ul')[1].get_text())
    else:
        sp_vacant_positions.append(soup.select('.pcontent ul')[0].get_text())

    #How to Apply
    sin_pos_how_to_apply = soup.select('.pcontent ol')[0]
    sp_how_to_apply_ol.append(sin_pos_how_to_apply)


    #Lower Section
    for sin_para in soup.select('.pcontent h5 ~ p'):
        sin_para_text = sin_para.get_text()

        if "Last Date" in sin_para_text:
            only_last_date = sin_para.get_text()
            sp_last_date.append(only_last_date)
        elif "To Apply Online" in str(sin_para):
            only_apply_link = sin_para.select('a')[0].get('href')
            sp_online_apply_link.append(only_apply_link)
        elif "Address" in str(sin_para):
            only_address = sin_para.get_text()
            sp_address.append(only_address)
        elif "Advertisement Image" in sin_para_text:
            only_link_ad_img = sin_para.select('a')[0].get('href')
            sp_advertisement_image_link.append(only_link_ad_img)
        elif "Application Form Will" in sin_para_text:
            sp_application_form_link.append("Application Form Will be Available Sn")
        elif "Application Form" in sin_para_text:
            only_app_form_link = sin_para.select('a')[0].get('href')
            sp_application_form_link.append(only_app_form_link)
        else:
            #print(sin_para)
            print(".")


    if not sp_last_date:
        sp_last_date.append("Available Soon")
    if not sp_online_apply_link:
        sp_online_apply_link.append("Available Soon")
    if not sp_address:
        sp_address.append("Available Soon")
    if not sp_application_form_link:
        sp_application_form_link.append("Available Soon")

    if not sp_advertisement_image_link:
        sp_advertisement_image_link.append("Available Soon")
    else:
        #Opening Next Page for Ad Image
        inner_response = requests.get(only_link_ad_img)
        innerhtmlcontent = inner_response.content
        innersoup = BeautifulSoup(innerhtmlcontent, 'html.parser')
        #Next Page Ad Image
        inner_ad_img = innersoup.select('.pcontent img')
        sp_inner_ad_img.append(inner_ad_img[0].get('data-lazy-src'))
    
    if not sp_inner_ad_img:
        sp_inner_ad_img.append("Not Available")

    single_post_list_wrapper = {'company_logo':sp_company_logo,
                                'main_title':sp_main_title,
                                'location':sp_location,
                                'vacancies':sp_vacancies,
                                'availability':sp_availability,
                                'main_experience':sp_main_experience,
                                'main_region':sp_main_region,
                                'main_education':sp_main_education,
                                'job_title':sp_title,
                                'categories':sp_categories,
                                'salaries':sp_salaries,
                                'age':sp_age,
                                'gender':sp_gender,
                                'designation':sp_designation,
                                'vacant_positions':sp_vacant_positions,
                                'how_to_apply':sp_how_to_apply_ol,
                                'last_date':sp_last_date,
                                'online_apply_link':sp_online_apply_link,
                                'address':sp_address,
                                'advertisement_img_link':sp_advertisement_image_link,
                                'app_form_link':sp_application_form_link,
                                'inner_ad_img':sp_inner_ad_img
                                }
    
    #print(single_post_list_wrapper)

    #Create DataFrame from list
    single_post_entry_df = pd.DataFrame(single_post_list_wrapper)
    return single_post_entry_df


def extract_links(main_page_url):
    main_response = requests.get(main_page_url)
    main_html_content = main_response.content
    main_soup = BeautifulSoup(main_html_content, 'html.parser')
    #Collect All Data of one post
    final_list_to_merge = []
    for sin_tr in main_soup.select('table.table-hover-jobs tr'):
        if sin_tr.find('th'):
            continue 
        # Extracting post title
        job_title = sin_tr.select('td:nth-child(2) a')
        url = job_title[0].get('href')
        try:
            final_list_to_merge.append(single_post_fetch(url))
            print(url+" Done..!")
        except:
            print(url+" Error....!")
    final_df = pd.concat(final_list_to_merge, ignore_index=True)
    return final_df
    



f_final_df = extract_links('https://jobsalert.pk/all-jobs/page/2')
f_final_df
#Replacing Same Value Patterns from DataFrame
#final_df = pd.concat(final_list_to_merge, ignore_index=True)



# f_final_df['vacancies'].replace(to_replace ='Vacancy', value = '', regex = True, inplace=True)
# f_final_df['availability'].replace(to_replace ='Availability', value = '', regex = True, inplace=True)
# f_final_df['main_experience'].replace(to_replace ='Experience[\n]', value = '', regex = True, inplace=True)
# f_final_df['main_region'].replace(to_replace ='Region', value = '', regex = True, inplace=True)
# f_final_df['main_education'].replace(to_replace ='Education', value = '', regex = True, inplace=True)
# f_final_df['main_education'].replace(to_replace ='\n', value = ' ', regex = True, inplace=True)
# f_final_df['age'].replace(to_replace ='\n', value = ' ', regex = True, inplace=True)
# f_final_df['salaries'].replace(to_replace ='\n', value = ' ', regex = True, inplace=True)
# f_final_df['designation'].replace(to_replace ='\n', value = ' ', regex = True, inplace=True)
# f_final_df['vacant_positions'].replace(to_replace ='\n', value = ' ', regex = True, inplace=True)
# f_final_df.to_csv('new.csv')



#Clear Excel Formatting
#pd.io.formats.excel.ExcelFormatter.header_style = None
#Exporting to Excel
#final_df.to_excel("new.xlsx")


```
